````python
data = []

for i in range(5):
  elements = str(input())

  data = autoAdd(data, elements)

print(data)

#1,3,4,2/apnd
#1,a/ext
#0/apnd
#b,ccasd/apnd
#1,2,3,4,5,6,7,8,9/apnd
````
